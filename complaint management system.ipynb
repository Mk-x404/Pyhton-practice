```python
# NOTE: "Complaint Management System"
# ðŸ”¹ 1. Submit a Complaint 

def submitted_complaints(a):
    with open("complaints.txt", "a") as file:
        file.write(a + "\n")
        print("Your complaint has been submitted successfully.")
                    
    with open("complaints.txt", "r") as file:
        content = file.readlines()
        content = [line.strip() for line in content if line.strip()] #clean them
        return content
        
def view_complaints():
    with open("complaints.txt", "r") as file:
        content = file.readlines()
        content = [line.strip() for line in content if line.strip()]  # Remove empty lines and strip spaces
        return content
            
def remove_complaints(s):
    with open("complaints.txt", "r") as file:
        content_read = file.readlines()
        clean_complaint = [line.strip() for line in content_read if line.strip()]  # Clean them

    # Remove the complaint by index 
    try :    
        #! Check if the index is valid
        index = int(s) - 1 
        if index < 0 or index >= len(clean_complaint):
            print("Invalid complaint number. Please try again.")
            return clean_complaint  # Return unchanged list
        

        delete_complaint = clean_complaint.pop(index).strip()  # Remove from file content
        
        print( f"Complaint '{delete_complaint.strip()}' has been removed successfully.")
        
        # Save updated list
        with open("complaints.txt", "w") as file:
            for line in clean_complaint:
                file.write(line + "\n")
                
            return clean_complaint  # Return updated list
            
    except Exception as e:
        print("Something went wrong while removing the complaint:", e)
        return None
    
    
while True:
    
    menu = ["Submit a Complaint", "View Complaints", "Remove a Complaint", "Exit"]
    for i, option in enumerate(menu, start=1):
        print(f"{i}. {option}")
    
    choice = input ("Please choose an option (1-4): " + ", ".join(menu) + ": ")
    if choice.isdigit():  # Check if input is a number (like "1", "2", etc.)
        choice_num = int(choice)  # Convert to number for range checking
        if choice_num < 1 or choice_num > 4:
            print("Please enter a number between 1 and 4.")
            continue
    else:
        print("That's not a valid number. Please enter 1, 2, 3, or 4.")
        continue
    
    if choice == "4":
        print("Exiting the Complaint Management System. Goodbye!")
        break
        
    # Submit a Complaint
    if choice == "1" :
        print("Welcome to the Complaint Management System!")
        user = input("Please enter your complaint: ")
        submit = submitted_complaints(user)
        
        for i , option in enumerate (submit):
            print (f"{i+1}. {option}")
    
    
    # View Complaints
    elif choice == "2":
        print("Here are the complaints submitted so far:")
        view = view_complaints()
        for i, display in enumerate(view, start=1):
            print (f"{i}. {display}")
        
# Remove a Complaint
    elif choice == "3":
        user_rem = input("Enter the number of the complaint you want to remove: ")
        complaint = remove_complaints(user_rem)
    #? Display updated list of complaints
        print("Updated list of complaints:")
        for i, submitted_complaint in enumerate(complaint, start=1):
            print(f"{i}. {submitted_complaint.strip()}")
        
        
#?  End of Complaint Management System

 
```
